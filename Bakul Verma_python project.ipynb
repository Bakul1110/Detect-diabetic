What is an IMU?

An inertial measurement unit (IMU) measures and reports raw or filtered angular rate and specific force/acceleration experienced by the object it is attached to. Data outputs for an IMU are typically body-frame accelerations, angular rates, and (optionally) magnetic field measurements.

The columns of the dataset are:

1. Category: The category to which this IMU belongs.

2. Manufacturer: Name of the manufacturer.

3. part number: Discrete part number

4. sku: stock keeping unit

5. alt_sku: alternate stock keeping unit

6. Description: description of the product

7. Status: Whether it is in stock or not

8. Quantity: Quantity available

9. Price: Price of the product

10. sku_create_date: when was the product procured

# 7. Ways to prevent Aging of products

This is basically referred to as Inventory Aging.


Aging inventory is any item that sits in your warehouse and doesn’t sell either quickly or at the full retail price. 


The age analysis always starts with the receiving date, meaning when an item is added to inventory.


Following ways can be adopted to prevent it:
   
   1). Eliminating obsolete inventory- Look for the product which is no longer in demand and reduce or stop the production
    of that particular product.
   
   2). provide discounts if the aging product is facing competition from the similar products existing in the market.
   
   3). Keep an eye over the condition of product. check if they are returned ones due to their malfunctioning.
   
   4). Maintain the quality of the product.
    
   5). Last but not the least(Maintain regular Inventory health analysis)
